```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

kickstarter_cleaned = pd.read_csv('kickstarter_cleaned_index.csv')
new_kickstarter_cleaned = kickstarter_cleaned.drop(columns='Unnamed: 0')
```

```python
new_kickstarter_cleaned.shape
```

```python
#leave the data only with "failed" and "successful"
failed = new_kickstarter_cleaned[new_kickstarter_cleaned.state == "failed"]
success = new_kickstarter_cleaned[new_kickstarter_cleaned.state == "successful"]
new_kickstarter_cleaned2 = pd.concat([failed,success])
new_kickstarter_cleaned2
```

```python
#country
country = new_kickstarter_cleaned2.groupby(['country']).state.value_counts(normalize=True).unstack()
print(country)
country.plot(kind='bar',stacked=True)
plt.title('Frequency Counts of Failed and Successful Projects by Country')
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', ncol=1)
plt.grid()
plt.show()
```

```python
#country
sns.countplot(new_kickstarter_cleaned2.country)
```

```python
#backers
#new_kickstarter_cleaned2.backers
backers = new_kickstarter_cleaned2.groupby(['backers']).state.value_counts(normalize=True).unstack()
backers
backers.plot(kind='bar',stacked=True)
```

```python
#backers
sns.stripplot(x="state", y="backers", data=new_kickstarter_cleaned2, jitter=True) 
```

```python
new_kickstarter_cleaned2['length_in_days']=new_kickstarter_cleaned2.project_length.str[0:2]#unit in days
#plot in percentage
C = new_kickstarter_cleaned2.groupby(['length_in_days']).state.value_counts(normalize=True).unstack()
print(C)
print(C.plot(kind='bar',stacked=True))
#plot without percentage
C = new_kickstarter_cleaned2.groupby(['length_in_days']).state.value_counts().unstack()
print(C)
print(C.plot(kind='bar',stacked=True))
```

```python
#usd_pledge_real
new_kickstarter_cleaned2.usd_pledged_real
D = new_kickstarter_cleaned2.groupby(['country']).usd_pledged_real.mean()
D.plot(kind='bar')
plt.title('Avgerage Pledged Amount by Country')
plt.show()
```

```python
#usd_pledge_real
new_kickstarter_cleaned2.usd_pledged_real
E = new_kickstarter_cleaned2.groupby(['country']).usd_pledged_real.median()
E.plot(kind='bar')
plt.title('Median Pledged Amount by Country')
plt.show()
```

```python
#usd_pledge_real
new_kickstarter_cleaned2.usd_pledged_real
F = new_kickstarter_cleaned2.groupby(['country']).usd_pledged_real.std()
F.plot(kind='bar')
plt.title('SD of Pledged Amount by Country')
plt.show()
```

```python
#usd_goal_real
G = new_kickstarter_cleaned2.groupby(['country']).usd_goal_real.mean()
G.plot(kind='bar')
plt.title('Avgerage Goal Amount by Country')
plt.show()
```

```python
#usd_goal_real
G = new_kickstarter_cleaned2.groupby(['country']).usd_goal_real.median()
G.plot(kind='bar')
plt.title('Median Goal Amount by Country')
plt.show()
```

```python
#usd_goal_real
G = new_kickstarter_cleaned2.groupby(['country']).usd_goal_real.std()
G.plot(kind='bar')
plt.title('STD Goal Amount by Country')
plt.show()
```
