```python
import pandas as pd
kickstarter = pd.read_csv('ks-projects-201801.csv')
```

```python
print(kickstarter[kickstarter.duplicated()]) #no completely duplicated rows
```

```python
kickstarter.columns
```

```python
#print(kickstarter.launched.sort_values())
index_A = kickstarter.launched.sort_values()[0:7].index
```

```python
index_B = kickstarter[kickstarter.name.isna()].index
```

```python
success = kickstarter[kickstarter['pledged'] > kickstarter['goal']] 
wrong = success[success['state']=='failed']
print(wrong.head())
index_C = wrong.index
```

```python
undefined = kickstarter[kickstarter['state'] == 'undefined']
print(undefined.country.unique())
index_D = kickstarter[kickstarter.country == 'N,0"'].index
```

```python
#kickstarter.deadline.sort_values() looks ok
kickstarter['deadline'] = pd.to_datetime(kickstarter.deadline)
kickstarter['launched'] = pd.to_datetime(kickstarter.launched)
kickstarter['project_length'] = kickstarter['deadline'] - kickstarter['launched']
kickstarter.project_length.sort_values() #no negative days, so should be fine
allindexes_remove = list(index_A) + list(index_B) + list(index_C) + list(index_D)
len(allindexes_remove) == len(set(allindexes_remove)) #all unique
allindexes_remove.sort()
allindexes_remove
```

```python
kickstarter_cleaned = kickstarter.drop(labels=allindexes_remove)

#LIST OF ROWS REMOVED:
#REMOVED ROWS WITH STRANGE LAUNCH DATE (1970)
#REMOVED ROWS WITH NA FOR NAMES
#REMOVED ROWS WITH PROJECTS THAT MADE GOAL BUT CONSIDERED FAILED
#REMOVED ROWS WITH COUNTRY 'N,0"'
kickstarter_cleaned.to_csv('kickstarter_cleaned_index')
```
