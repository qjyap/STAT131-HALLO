```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
kickstarter_cleaned = pd.read_csv('kickstarter_cleaned_index.csv')
new_kickstarter_cleaned = kickstarter_cleaned.drop(columns='Unnamed: 0')
```

Extremely Successful Kickstarters

```python
new_kickstarter_cleaned['percent_funded'] = new_kickstarter_cleaned['usd_pledged_real'] / new_kickstarter_cleaned['usd_goal_real']
new_kickstarter_cleaned['percent_funded'].describe()
```

```python
extremelyFunded = new_kickstarter_cleaned['percent_funded'] > 10
new_kickstarter_cleaned.sort_values(by = 'percent_funded',ascending = False)
```

```python
extremelyFundedKickstarters = new_kickstarter_cleaned[extremelyFunded]

extremelyFundedKickstarters.head()
```

```python
extremelyFundedKickstarters['usd_goal_real'].value_counts()
```

```python
extremelyFundedKickstarters[['usd_goal_real', 'percent_funded']].sort_values(by = 'percent_funded', ascending = False)
```

```python
extremelyFundedKickstarters['usd_goal_real'].value_counts().sort_index().plot()
```

```python
extremelyFundedKickstarters[extremelyFundedKickstarters['usd_goal_real'] <= 120000]['usd_goal_real'].value_counts().sort_index().plot()
```

```python
extremelyFundedKickstarters['usd_goal_real'].describe()
```

```python
extremelyFundedKickstarters[extremelyFundedKickstarters['usd_goal_real'] <= 10000]['usd_goal_real'].describe()
```

```python
extremelyFundedKickstarters.groupby('main_category')['percent_funded'].describe()
```

Average Pledge

```python
avgPledgeByCountry = new_kickstarter_cleaned.groupby('country')['pledged_per_backer']
avgPledgeByCountry.mean().plot(kind = 'bar', stacked = True)
plt.title('Average Pledge Amount per Country')
plt.show()
```

```python
avgPledgeByMainCategory = new_kickstarter_cleaned.groupby('main_category')['pledged_per_backer']
avgPledgeByMainCategory.mean().plot(kind = 'bar', stacked = True)
plt.title('Average Pledge Amount per Main Category')
plt.show()
```
