```python
import pandas as pd
kickstarter = pd.read_csv(r"ks-projects.csv")
```

```python
print(kickstarter.sort_values(by = 'launched')[['name', 'launched', 'deadline']])
print(kickstarter.sort_values(by = 'deadline')[['name', 'launched', 'deadline']])
```

```python
print(kickstarter[kickstarter['launched'].isna()])
print(kickstarter[kickstarter['deadline'].isna()])
```

```python
print(kickstarter[kickstarter['name'].duplicated(keep = False)].sort_values(by = 'name'))
print(kickstarter[kickstarter['ID'].duplicated(keep = False)].sort_values(by = 'ID'))
```
