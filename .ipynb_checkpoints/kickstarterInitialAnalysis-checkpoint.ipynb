```python
import pandas as pd
kickstarter = pd.read_csv('ks-projects-201801.csv')
```

```python
print(kickstarter.isna().sum()) #4 missing values for name, 3797 missing for usd pledged (we will be disregarding that column, will be explaining this)
print(kickstarter.shape)
```

```python
print(kickstarter['ID'].nunique())
print(kickstarter['name'].nunique()) #strange, ~3000 rows seem to have duplicated names with unique IDs

print(kickstarter[kickstarter['name'].duplicated(keep = False)].sort_values(by = 'name'))
print(kickstarter[kickstarter['ID'].duplicated(keep = False)].sort_values(by = 'ID'))
```

```python
print(kickstarter[kickstarter['name'].isnull()]) #these are the rows with no name
print(kickstarter[kickstarter['name'].isnull()].index) #obtaining the index of those rows for deletion
```

```python
print(kickstarter['category'].nunique()) #159 categories provided. These are not all main categories, most are subcategories 
print(kickstarter['main_category'].nunique()) #15 main categories. Aligns with Kickstarter's categories on website.
print(kickstarter['category'].unique()) #looks great
print(kickstarter['main_category'].unique()) #Nothing out of the ordinary
```

```python
print(kickstarter.sort_values(by = 'launched')[['name', 'launched', 'deadline']])
print(kickstarter.sort_values(by = 'deadline')[['name', 'launched', 'deadline']])
```
